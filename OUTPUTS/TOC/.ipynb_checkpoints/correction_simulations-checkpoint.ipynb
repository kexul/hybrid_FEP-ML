```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import pandas as pd 
pd.set_option('display.max_rows', 10)
import scipy
from sklearn import metrics
import time
from scipy.stats import gaussian_kde
```

```python
def rescale_data(data, original):
    """
    rescale an array according to another array (original)
    """
    data_array = np.array(data)
    scaled_data = np.interp(
                            data_array, 
                            (data_array.min(), data_array.max()), 
                            (min(original), max(original)) 
                            )
    return list(scaled_data)
def permutate_value(value, deviate):
    """
    permutate a value randomly in a scale of deviate variable
    """
    # generate a random number within the range:
    factor = np.random.uniform(-deviate, deviate)
    # permutate the FEP value to generate a 'slightly off' prediction
    random_pred = value - factor

    return random_pred
```

```python
# read in dataset
dataset = pd.read_csv("dataset_source.txt", sep=";", usecols=[0,3,5], names=["cpd", "exp", "fep"])[1:].sample(15)
dataset
```

```python
# take arrays of the data:
original_exp = dataset["exp"].astype(float).values
original = dataset["fep"].astype(float).values

# generate a worse prediction of the experimental values by transforming the fep values with noise:
worse = [ permutate_value(value, 8) for value in original ]
worse = rescale_data(worse, original)
```

```python
# make base plot with annotations:
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# plot the noise prediction:
axes[0].scatter(worse, original_exp)


# generate a better prediction and plot:
improved_prediction = []
for pred, exp in zip(worse, original_exp):
    average = (pred + exp + exp)/3
    improved_prediction.append(average)
axes[1].scatter(improved_prediction, original_exp)
    
    
for ax in axes:
    # set limits:
    ax.set_ylim(-12, 0)
    ax.set_xlim(-12, 0)
    
    # make diagonals:
    ax.fill_between(
                    x=[-25, 0], 
                    y2=[-24.5,0.5],
                    y1=[-25.5,-0.5],
                    lw=0, 
                    zorder=-10,
                    alpha=0.5,
                    color="darkorange")
    # upper bound:
    ax.fill_between(
                    x=[-25, 0], 
                    y2=[-24,1],
                    y1=[-24.5,0.5],
                    lw=0, 
                    zorder=-10,
                    color="darkorange", 
                    alpha=0.2)
    # lower bound:
    ax.fill_between(
                    x=[-25, 0], 
                    y2=[-25.5,-0.5],
                    y1=[-26,-1],
                    lw=0, 
                    zorder=-10,
                    color="darkorange", 
                    alpha=0.2)

    # remove axis info:   
    ax.axes.get_yaxis().set_visible(False)
    ax.axes.get_xaxis().set_visible(False)
    
# add textual information:
plt.tight_layout(w_pad=9)    
axes[0].set_title("Standalone FEP", fontsize="x-large")
axes[1].set_title("FEP+ML hybrid", fontsize="x-large")

axes[0].text(s=r"Experimental FE / kcal mol$^{-1}$", x=-13, y=-10, rotation=90, fontsize="x-large")
axes[0].text(s=r"Predicted FE / kcal mol$^{-1}$", x=-6, y=-13, fontsize="x-large", ha="center")
axes[1].text(s=r"Predicted FE / kcal mol$^{-1}$", x=-6, y=-13, fontsize="x-large", ha="center")


axes[0].arrow(x=0.5, y=-6, dx=2, dy=0, clip_on=False, head_width=0.5, color="black", overhang=0.8)
axes[0].text(s="Plug in\n\ncorrection\nterms", x=1.85, y=-5, style='oblique', ha='center',
         va='top', wrap=True, fontsize="x-large")

# add correction term arrows:
for pred, corrected_pred, exp in zip(worse, improved_prediction, original_exp):
    predicted_offset = corrected_pred - pred
    if abs(predicted_offset) > 4:
        axes[0].arrow(x=pred, y=exp, dx=predicted_offset, dy=0, overhang=0.8, head_width=0.3)
        axes[0].text(s="ML-predicted\n$\Delta$G$_{offset}$ value", x=pred+2.7, y=exp+0.8, style='oblique', ha='center',
         va='top', fontsize="x-large")
    else:
        axes[0].arrow(x=pred, y=exp, dx=predicted_offset, dy=0, overhang=0.8, head_width=0.3, alpha=0.3)


        
        
plt.savefig("process_subplots.png", dpi=300, bbox_inches="tight")
plt.show()
```
